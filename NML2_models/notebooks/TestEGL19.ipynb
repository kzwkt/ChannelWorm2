```python
# First run ../analyse_EGL19.sh

from pyneuroml import pynml

egl19datafile = '../../parameter_fitting/basic_fitter/data/EGL-19_patch_clamp.csv'

vs = []
iss = []
with open(egl19datafile) as f:
    for l in f:
        w = l.split(',')
        vs.append(float(w[0]))
        iss.append(float(w[1]))
        

data, indeces = pynml.reload_standard_dat_file('../ca_boyle.i_peak.dat')

print('Loaded data: %s'%data.keys())

scale = 30
xs = [vs,[v*1000 for v in data['t']]]
ys = [iss, [scale*i*1e12 for i in data[0]]]

markers = ['o',None]
linewidths = [0,2]
labels = ['data','NeuroML']

pynml.generate_plot(xs, 
                   ys, 
                   'Peak currents',
                    labels = labels,
                    markers = markers,
                    linewidths = linewidths,
                   grid=True)
```
